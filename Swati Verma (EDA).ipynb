## 1. Explain steps for Data Processing? Replace Processing with Analysis


ANSWER:-

Here are the steps for Data Analysis:

1. Data Collection: The first step in data analysis is to collect the relevant data from various sources. This data may be in different formats such as structured or unstructured data.

2. Data Cleaning: The next step is to clean the data. This involves removing any irrelevant or duplicate data and correcting any errors in the data.

3. Data Integration: Once the data has been cleaned, it needs to be integrated from various sources into a single dataset that can be used for analysis.

4. Data Transformation: Data transformation involves converting the raw data into a format that is suitable for analysis. This may involve standardizing units of measurement or converting categorical data into numerical data.

5. Data Analysis: Once the data has been transformed, it is ready for analysis. There are various techniques that can be used to analyze the data, such as statistical analysis, machine learning, and data visualization.

6. Interpretation and Communication of Results: The final step in data analysis is to interpret the results and communicate them to stakeholders. This involves presenting the findings in a way that is easily understandable and actionable.

It's worth noting that the above steps can also be applied to data processing, with the primary difference being that the end goal of data processing is to prepare data for analysis, while the end goal of data analysis is to gain insights and make informed decisions based on the data.

## 3. Explore the Concept of Handling Data Missing with help of example.

Handling missing data is an important aspect of data analysis. In Python, we can use the Pandas library to handle missing data. Let's look at an example to understand how to handle missing data.

Suppose we have a dataset that contains information about the employees in a company. 

```python
employee_data= pd.read_excel('C:\\Users\\HP\\Desktop\\DS-ML-IOT\\table data/Employee_data.xlsx')
employee_data
```

```python
employee_data.isnull().sum()
```

As we can see from above data tha we have 2 missing value in age column, 4 in salary column, 2 in department and 3 in year of experience. There are different method for treating the missing value of different data type. In this data we have catagorial and numerical data.

```python
Numerical_data= [i for i in employee_data if employee_data[i].dtype!=object]
data1=employee_data[Numerical_data]
data1
```

Above data show the columns with numerical data.

Now let see different step to fill or remove the missing values.

1. Drop missing values:
One way to handle missing data is to simply drop the rows or columns that contain missing values. We can use the dropna() function to do this. For example, to drop all rows with missing values, we can do:

```python
droping= data1.dropna()
droping
```

In the droping table we have remove all the row which is having missing values.

2. Fill missing values:
Another way to handle missing data is to fill in the missing values with some appropriate values. We can use the fillna() function to do this. For example, to fill in the missing values in the Salary column with the median salary, we can do:

```python
employee_data['Salary'].fillna(employee_data['Salary'].median(),inplace= True)
employee_data.isnull().sum()
```

3. Interpolate missing values:
Another way to handle missing data is to interpolate the missing values based on the values of the other rows. We can use the interpolate() function to do this. For example, to interpolate the missing values in the Age column, we can do:

```python
employee_data['Age'].interpolate(inplace=True)
employee_data
```

From above table we can see that our age and salary columns are completely fill with the values

## 4. Explain about Univariate, Bivariate and Multivariate Analysis?

Univariate, bivariate, and multivariate analysis are three common types of statistical analysis used in research and data analysis. Each type of analysis examines different aspects of a data set and can provide insights into different types of relationships between variables. Here's a brief explanation of each. For instance, univariate analysis is used to examine the distribution of a single variable, bivariate analysis is used to identify the relationship between two variables, and multivariate analysis is used to explore the relationships between three or more variables.

1. Univariate analysis: This is the simplest type of analysis and involves analyzing a single variable at a time. Univariate analysis describes the distribution of a single variable, such as the mean, median, and mode. It also describes measures of variability such as the standard deviation. This type of analysis is often used to get an initial understanding of the data and can help identify patterns, outliers, or trends. The most common type of univariate analysis is to describe a variable using summary statistics, which include measures of central tendency (e.g. mean, median, mode) and measures of variability (e.g. standard deviation).

2. Bivariate analysis: Bivariate analysis examines the relationship between two variables. It is used to determine whether there is a correlation or association between the two variables. Bivariate analysis can be done using various statistical techniques such as correlation analysis, t-tests, chi-square tests, and regression analysis. For example, if we want to determine the relationship between height and weight of a group of people, we would conduct bivariate analysis. Bivariate analysis is a correlation that measures the strength of association between two variables and the direction of the relationship.

3. Multivariate analysis: Multivariate analysis involves the simultaneous analysis of three or more variables. It is used to determine how multiple variables are related to each other. It can be utilized to identify patterns or relationships that are not apparent in bivariate analysis. Multivariate analysis can be done using techniques such as factor analysis, cluster analysis, and multiple regression analysis. In order to understand the factors influencing a person's income, we may use multivariate analysis to examine variables such as education level, occupation, and location. Median household income is calculated by finding the middle number in a group of numbers. For example, if a household has three incomes of Rs35000,Rs40000, and Rs45,000. The median income would be Rs45,500. The Census Bureau also counts households with no income in its calculation.

## 5. What is meant by standard deviation and variance for a numerical data?

Standard deviation and variance are both measures of how spread out a set of numerical data is. They describe a dataset's variability or dispersion. The standard deviation is a measure of how spread out a set of numerical data is, calculated by finding the mean of the data points, subtracting the mean from each data point, squaring the difference of each result, finding the mean of those squared differences, and then taking the square root of the mean.

The variance is a measure of how much the data deviate from the mean. It is calculated by taking the average of the squared differences between each data point and the mean of the dataset. A high variance indicates that the data is spread over a wide range of values. In contrast, a low variance indicates that the data clusters more closely around the mean.

The standard deviation is the square root of variance. It measures the amount of variation or dispersion in a set of data. It tells us how far the data points are from the mean on average. Like variance, a high standard deviation indicates that the data is more spread out. In contrast, a low standard deviation indicates that the data is tighter clustered around the mean.

It is a well known fact that variance and standard deviation are both important statistical measures that are used in many different fields of research and data analysis, such as finance, engineering, sociology, and so on. The information they provide can help us understand the characteristics of a dataset and make informed decisions.

## 6. What is meant by imbalance in the categorical data and how do we handle that?

Imbalance in categorical data refers to a situation where one category or class has significantly more observations than the others. This is a common problem in machine learning, where algorithms trained on imbalanced data can become biased towards the majority class, leading to poor performance in predicting the minority class. To address this issue, it is important to balance the dataset by either oversampling or undersampling the categories. By doing this, the machine learning algorithm is able to generalize better, leading to better performance.

## 7. What are various ways to remove skew in the numerical data?


Skewness is a measure of the asymmetry of a numerical data distribution. Positive skewness means that the tail of the distribution is on the right side and negative skewness means that the tail of the distribution is on the left side. In order to remove skewness from numerical data, there are several techniques used:

1. Log Transformation: Log transformation transforms data into a logarithmic scale. This helps reduce outlier impact and normalizes the data. 

2. Square Root Transformation: Square root transformation reduces data skew. It is particularly useful for data with a long tail on the right side. 

3. Box-Cox Transformation: Box-Cox transformation transforms data into normal distribution. It is a family of transformations that includes log transformation, square root transformation, and others.

4. Winsorization: Winsorization is a technique that replaces extreme values in the data with less extreme values. This helps to reduce the impact of outliers on the data.

5. Z-Score Transformation: Z-Score transformation transforms data into a standard normal distribution. This helps reduce outlier impact and normalizes the data.

6. Quantile Transformation: Quantile transformation transforms data into a uniform distribution. This helps reduce outlier impact and normalizes the data.

7. Robust Scaling: Robust scaling scales the data using the interquartile range instead of the standard deviation. This helps to reduce the impact of outliers on the data.

8. Kernel Density Estimation: Kernel density estimation estimates the data density. This helps reduce outlier impact and normalizes the data.

## 8. Explain IQR method to remove outliers in the numerical data?

The IQR (interquartile range) method is used to identify and remove outliers from numerical data. It is a robust method that is less sensitive to extreme values than other methods such as mean and standard deviation.

The IQR is defined as the range between the first quartile (Q1) and the third quartile (Q3) of the data. The first quartile is the value below which 25% of the data falls and the third quartile is the value below which 75% of the data falls. The IQR is calculated as IQR = Q3 - Q1.

To identify outliers using the IQR method, the data is first sorted ascending. Then, the lower and upper bounds are calculated as follows:

- Lower Bound = Q1 - (1.5 x IQR)

- Upper Bound = Q3 + (1.5 x IQR)

Any data point that falls below the lower bound or above the upper bound is considered an outlier and can be removed from the data.

It is imperative to note that the choice of 1.5 as the multiplier for the IQR can be adjusted depending on the nature of the data. The degree to which outliers can be tolerated is also determined by the outlier tolerance. A higher multiplier will result in a larger range of accepted values, while a lower multiplier will result in a smaller range of accepted values.

The IQR method can be applied to univariate data, where only one variable is considered at a time. It can also be applied to multivariate data, where multiple variables are considered simultaneously. In the latter case, outliers can be identified based on their distance from the center of the data using techniques such as Mahalanobi's distance.
